```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


imagem = cv2.imread('/content/drive/MyDrive/BCC/10/RP/2VA/Flood Area Segmentation/Image/1003.jpg', cv2.IMREAD_GRAYSCALE)


if imagem is None:
    print('No foi possvel carregar a imagem.')
else:
    # segmentao de Otsu
    _, imagem_segmentada = cv2.threshold(imagem, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Exibi a imagem
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(imagem, cmap='gray')
    plt.title('Imagem Original')

    plt.subplot(1, 2, 2)
    plt.imshow(imagem_segmentada, cmap='gray')
    plt.title('Imagem Segmentada')

    plt.show()
```
